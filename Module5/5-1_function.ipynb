```python
# 建立 function 1
# 某個物件才能使用的function稱為method，如list.append()

def greeting():
    print("Hello World!")

greeting()
```

```python
# 建立 function 2: 有return值
def get_greeting_str():
    msg = "Hello" + " " + "World"
    return msg

#令一變數去接他
result = get_greeting_str()
print(result)
```

```python
# 建立 function 3: 有參數

def greeting(name):
    print(f"Hello, , {name}!")

greeting("Iris")
```

```python
# 建立 function 4 -> 預設參數值(greeting = "Hello")一定要擺在後面

def greeting(name, greeting = "Hello"):
    print(f"{greeting} , {name}")

greeting("Iris")
greeting("Iris", "Good Morning")
greeting(greeting = "Good Morning", name = "Iris")
```

```python
def add(*arg, **kwargs):
    print(kwargs)
    total = 0
    for i in arg:
        total += i
        return total

total = add(2,5,7,3,1,7, unit = "元")
print(total)
```
