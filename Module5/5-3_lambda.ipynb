### lambda 匿名函式  
方便把簡短的function寫成一行

`function_name = lambda arguments: expression`

````python
# 建立函式
def add(x, y):
    return x + y
result = add(10,5)
print(result)
````

````python
# 基本用法
add = lambda x, y: x + y
result = add(5,6)
print(result)
````

````python
# 條件判斷
def abs(x):
    if x > 0:
        return x
    else: 
        return -x
result = abs(-10)
print(result)
````

````python
# 三元運算 
# <if 為true時想return的結果> if <判斷式> else <els要回傳的結果>
def abs(x):
    return x if x > 0 else -x

result = abs(-10)
print(result)

#用lambda寫: 
abs2 = lambda x: x if x > 0 else -x
result = abs2(-100)
print(result)
````

### 小試身手  
請建立一個 list ，並在該 list 內產生多個 `[1, 2, x]` 的 list (x 的值要遞增)  
x 到 10 即可  
Ex :  
```Python
[ [1, 2, 1], [1, 2, 2], [1, 2, 3]...[1, 2, 10] ]
```

````python
result = []
inside_list = [1,2,0]
for i in range(1, 11):
    inside_list[2] = i
    result.append(inside_list.copy())#若沒有使用.copy()，inside_list會直接被修改
print(result)
````

##### `sort()` 方法

````python
result = []
for i in range(1,11):
    result.append([1,2,i])
print(result)

# list.sort 預設由小到大
# 反序排列
result.sort(reverse=True)
print(result)
````

每三個 list 中間的值要改成 5  
Ex:  
```python
[ [1, 2, 1], [1, 2, 2], [1, 5, 3], [1, 2, 4], [1, 2, 5], [1, 5, 6]...]
```

````python
result = []

for i in range(1,11):
    if i % 3 == 0:
        result.append([1 ,5, i])
    else:
        result.append([1, 2, i])
print(result)

result.sort(reverse = True, key = lambda x: x[2])#看第三筆資料做排序的準則
print(result)
````
