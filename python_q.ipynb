### 有一筆學生資料如下:

```python
students_grade = [
    {"name": "Denji", "國文": 30, "英文": 10, "數學": 20},
    {"name": "Power", "國文": 60, "英文": 85, "數學": 70},
    {"name": "Aki", "國文": 95, "英文": 88, "數學": 92},
    {"name": "Pochita", "國文": 50, "英文": 65, "數學": 20},
    {"name": "Makima", "國文": 100, "英文": 98, "數學": 95},
    {"name": "Reze", "國文": 65, "英文": 40, "數學": 30},
]
```

##### 1. 印出所有學生姓名

```python
for student_name in students_grade:
    print(student_name["name"])
```

##### 2. 印出 `Pochita` 的數學成績

```python
for student_name in students_grade:
    if student_name["name"] == "Pochita":
        print(student_name["數學"])
```

#### 3. 計算每位學生的總分
輸出格式  
`Phochita 總分: xxx 分`

```python
for student_name in students_grade:
    print(f"{student_name['name']} 總分: {student_name['國文'] + student_name['英文'] + student_name['數學']}分")
```

#### 4. 計算每位學生的平均分數
輸出格式  
`Pochita 平均分數: xxx 分`

```python
for student_name in students_grade:
    print(f"{student_name['name']} 平均分數: {(student_name['國文'] + student_name['英文'] + student_name['數學']) / 3:.2f}分")
    # :.2f 將浮點數格式化為顯示兩位數的字串，進行四捨五入
```

#### 5. 找出數學最高分的學生

```python
math_highest = 0

for student_name in students_grade:
    if student_name["數學"] > math_highest:
        math_highest = student_name["數學"]
        math_highest_student = student_name["name"]
    
print(f"{math_highest_student} 考最高，數學成績為：{math_highest}分")
```

#### 6. 計算全班國文平均

```python
chinese_sum = 0
for student_name in students_grade:
    chinese_sum += student_name["國文"]
print(f"國文總分為：{chinese_sum}分")
print(f"國文平均分數為：{chinese_sum / len(students_grade):.2f}分")
```

#### 7. 找出不及格的學生（任一科 < 60）

```python
for student_name in students_grade:
    if student_name["英文"] < 60 or student_name["數學"] < 60 or student_name["國文"] < 60:
        print(f"{student_name['name']} 不及格")
```

#### 8. 新增「總分」到每個 dict
例如 :  
`{"name": "Pochita", "國文": 80, ..., "總分": 245}`

```python
for student_name in students_grade:
    student_name["總分"] = student_name["國文"] + student_name["英文"] + student_name["數學"]
    print(student_name)
```

#### 9. 找出三科都 > 80 的學生

```python
for student_name in students_grade:
    if student_name["英文"] >= 80 and student_name["數學"] >= 80 and student_name["國文"] >= 80:
        print(f"{student_name['name']} 優等生")
```

#### 10. 將資料轉成以下格式 (dict of dict)

```python
"""
{
    "Denji": {"國文": 30, "英文": 10, "數學": 20},
    .....
}
"""
student = {}
for student_name in students_grade:
    student[student_name["name"]] = {
        "國文": student_name["國文"],
        "英文": student_name["英文"],
        "數學": student_name["數學"]
    }
print(student)
```
