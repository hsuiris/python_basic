整數 (int)、浮點數 (float)

```python
num = 10
print(num)
float_num = 22.5
print(float_num)
```

布林值 (boolean)

```python
parse_ok = True
print(parse_ok)
parse_not_ok = False
print(parse_not_ok)
print(type(parse_not_ok))
```

字串 (string)

```python
name = "Jarvis"
print(name)
```

```python
text = "False"
print(text)
print(type(text)) #檢查變數的型別
```

列表 (list)

```python
nums = [1, 10, 20]
print(nums)
```

元組 (tuple)

```python
nums = (1, 2, 3)
print(nums)
```

集合 (set)

```python
nums = {1, 2, 3}
print(nums)
```

字典 (dict)

```python
user_info = {"name": "Jarvis", "age": 23}
print(user_info)
```

```python
#創建一個空的list
data = []
print(data)
print(type(data))
```

```python
# 創建一個空的 dict
info = {}
print(info)
print(type(info))
```

```python
# 創建一個空的 set 要使用set()
data = set()
print(data)
print(type(data))
```
