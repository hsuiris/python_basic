百貨公司周年慶 消費金額超過 2000 要折 200  

請撰寫一支程式  
計算客戶消費的結帳金額

```python
total = int(input())
if total >= 2000:
    total -= 200
print(total)
```

```python
# 滿2000折200, 滿4000折400, 滿6000折600

def discount(total):
    if total >= 2000:
        total -= 200*(total // 2000)
    return total

total = int(input())
print(discount(total))
```
