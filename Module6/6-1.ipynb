```python
# 基於需要哪些資料來使用底下的method才建立class
```

```python
# 類別名稱
class Car:
    # 類別屬性 (Class Attribute)
    wheels = 4
    
    # 建構函式
    def __init__(self, brand, model):
        self.brand = brand # 實例屬性(Instance Attribute) 
        self.model = model

    # 方法 (Method)
    def show_info(self):
        print(f"{self.brand} 有型號{self.model} 而且有{Car.wheels}個輪子")

#建立實例
car1 = Car("toyoto","IB43")
car1.show_info()
```
