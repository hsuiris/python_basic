```python
# 實做練習 將每個單字改成大寫開頭
string04 = "hello world! hello john"

data = string04.split()
# .split()把資料轉成list
print(data)

#看前兩筆資料
for i in data[:2]:
    print(i)

#創建一個空字串，以存放結果
result = ""

for i in data:
    result += i[0].upper() + i[1:] + " "

print(result)

#去除前後兩端空格
print(result.strip())
```

```python
string04 = "hello world! hello john"

data = string04.split()

for i in range(len(data)):
   data[i] = data[i][0].upper() + data[i][1:]
    #每個單字重新賦值
print(data)

#對迭代的物件做合併
result = " ".join(data)
print(result)
```
