現在很少用format
|語法|說明|
|:---:|:---:|
|%s |插入字串到字串中|
|%d |插入整數到字串中|
|%f |插入浮點數到字串中|
|%% |輸出百分比(類似跳脫字元)|

```python
# 多組文字
date1 = "Hello"
date2 = "World"

msg = "%s, %s!" %("Hello", "World")
print(msg)
```

```python
# 整數
age = 30
msg = "I'm %d years old" % (age)
print(msg)
```

```python
# 文字與整數
user = "Alex"
age = 15
msg = "%s is %d years old" % (user, age)
print(msg) 
```

```python
# 指定寬度 (維持 10 個字元長度，預設向右看齊)
data = "|%10s|" % "Hello"
print("ˍ"*12)
print(data)
print("¯"*12)
```

```python
# 靠左對齊 (維持 10 個字元長度，向左對齊)
data = "|%-10s|" % "Hello"
print("ˍ"*12)
print(data)
print("¯"*12)
```

```python
# 指定浮點數位數
# (維持10個字元長度，取到小數點後兩位)
data = "|%10.2f|" % 3.14567
print("ˍ"*12)
print(data)
print("¯"*12)
```

```python
# 指定文字長度上限 (只有文字才在格式化字串中加入「.」來限定字串長度)
# (只取3個字元的長度)
msg = "[%.3s]" % "Hello"
print(msg)
```

```python
# 空白補 0
data = "A118"
msg = "[%08.2s]" % data
print(msg)
```
