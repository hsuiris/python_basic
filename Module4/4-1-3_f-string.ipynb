```python
# f-string (又稱作 formatted string literals, version >= 3.6)
name = "Iris"
msg = f"My name is {name}"
print(msg)
```

```python
name = "Iris"
age = 24
msg = f"{name} is {age} years old"
print(msg)
```

```python
# 靠左對齊、靠右對齊
#預設靠左對齊，補上>即可靠右
name = "Iris"
result = f"Name: [{name:10}], Name: [{name:>10}]"
print(result)
```

```python
# 靠左對齊、靠右對齊，字數不足，補「X」
# (沒寫補什麼，預設空格)
name = "Iris"
result = f"Name: [{name:X<10}], Name: [{name:X>10}]"
print(result)
```

```python
# 置中對齊
name = "Iris"
result = f"Name: [{name:^10}]"
print(result)
```

```python
# 指定文字長度上限 (只有文字，才在格式化字串中加入「.」來限定字串長度)
name = "Iris"
result = f"Name: [{name:.2}]"
print(result)

#應用:身分證去識別化(只秀出前四碼)
user_id = "H426782268"
result = f"{user_id:X<10.4}"
print(result)
```
