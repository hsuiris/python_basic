```python
fruit_dict = {    
    'apple': {
        'price': 100,
        'quantity': 50,
        'color': ['green', 'red']
    },
    'orange': {
        'price': 50,
        'quantity': 30,
        'color': ['orange'] 
    }
}
print(fruit_dict)

#檢查某key是否在Dict裡面
print(fruit_dict.get("kiwi")) # 法一: 沒有該key會回傳None
print("apple" in fruit_dict) # 法二: dict搭配成員運算子
print("kiwi" in fruit_dict)
```

```python
# 使用 for 迴圈讀取 dict 的key -> dict.key()
for i in fruit_dict:
    print(i)
```

```python
# 使用 for 迴圈讀取 dict的value -> dict.value()
for value in fruit_dict.values():
    print(value)
```

```python
# 使用 for 迴圈讀取 dict 的item -> dict.item()

for i in fruit_dict.items():
    print(i)
    #輸出結果可看出類似tuple結構

# tuple裡面有兩個item會令兩個變數接出來
for key, value in fruit_dict.items():
    print(key, value)
```

```python
# 取得fruit_dict所有顏色
for key, value in fruit_dict.items():
    print(value["color"])
    for color in value["color"]:
        print(color)
```

```python
# 所有數量的加總
# 把key改成fruit, value改成info也可以，自由命名
sum = 0
for fruit, value in fruit_dict.items():
    print(value["quantity"])
    sum += value["quantity"]

print(f"總數量: {sum}")

```

```python
# 練習: 哪個水果價格最高，價格多少?

fruit_dict2 = {
    'apple': {
        'amount': 100,
        'color': ['green', 'red', 'yellow'],
        'price': 30
    },
    'kiwi': {
        'amount': 10,
        'color': ['yellow', 'green'], 
        'price': 150
    },
    'orange': {
        'amount': 50, 
        'color': ['orange'], 
        'price': 100
    },
    'tomato': {
        'amount': 80, 
        'color': ['red'], 
        'price': 50
    }
}

top_price = 0
top_fruit = "" 

for key, value in fruit_dict2.items():
    print(value["price"])
    if top_price < value["price"]:
        top_fruit = key
        top_price = value["price"]

print(f"最貴的水果: {top_fruit}, 價格: {top_price}元")
```

```python
# 資料處理: list裡面有資料重複，將之整理成Dict
fruit_color_list = [
    {'apple': 'red'}, 
    {'grape': 'purple'}, 
    {'kiwi': 'yellow'}, 
    {'apple': 'green'}, 
    {'kiwi': 'green'}, 
    {'grape': 'green'}]

result = {} #建立空Dict

for i in fruit_color_list:
    print(i)
    for fruit, color in i.items():
        if result.get(fruit) is None:
            # result.update(i) 原本寫這個，但是沒有把color變list，後面不能做append
            result[fruit] = [color] # !!!這個是我本來不會的。在result裡面建立這個key，放list[color]
        else: 
            result[fruit].append(color)

print(result)
```

##### `defaultdict(<type>)`

```python
# from 模組 import 函式/class
from collections import defaultdict
# 當訪問的鍵不存在時，可以自動創建鍵和預設值，而不是拋出KeyError。
# 當訪問的鍵存在時，可以自動加入


# 以int型態的預設值 0 來初始化
data = defaultdict(int)

print(data["apple"])
print(data)
```

```python
# 以list型態來初始化
result = defaultdict(list)
print(result["apple"])
print(result)
```

```python
# 前面的資料處理練習換種更有效的方式
fruit_color_list = [
    {'apple': 'red'}, 
    {'grape': 'purple'}, 
    {'kiwi': 'yellow'}, 
    {'apple': 'green'}, 
    {'kiwi': 'green'}, 
    {'grape': 'green'}]

result = defaultdict(list)

for fruit_color in fruit_color_list:
    for fruit, color in fruit_color.items():
        result[fruit].append(color) #呼叫到該key的同時創建一個空list，若本來有對應key的class會自動append

print(result)
```
