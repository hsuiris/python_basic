|string方法| 功能 |
|:---:|:---:|
|.replace() | 替換字串 |
|.strip() | 去除兩側空格 |
|.upper() | 字串小寫變大寫 |
|.lower() | 字串大寫變小寫 |

```python
# 字串變數初始化
name = "Alex"
print(name)
```

```python
# 替換字串
'''
str.replace(<old>, <new>)
將 string 中的 str1 替換成 str2
只有string會有.replace的用法
'''
name = "Alex"
new_name = name.replace("ex", "len") #開一個新的變數去接它
print(name) #不會改到原本的變數
print(new_name) 
```

```python
# 去除兩側空格
'''
str.strip()
去除字串 string 左、右兩邊的空格
'''
data = "   Pochita"
data = data.strip()
print(data)

data = "XPocXhitaXX"
print(data.strip("X"))
print(data.lstrip("X")) #去除左邊的。去除右邊: .rstrip()
```

```python
# 字串小寫變大寫
'''
str.upper()
將字串 string 裡的字母全部改成小寫
'''
data = "GOOD"
print(data.lower())

'''
str.lower()
將字串 string 裡的字母全部改成大寫
'''
data = "good"
print(data.upper())

'''
str.capitalize()
首字改成大寫
'''
data = "good"
print(data.capitalize())
```

```python
# str.split()
# 字串分割成 list
greeting = "hello world! hello John"
data = greeting.split()
print(data)

#若中間以X隔開
greeting = "helloXworld!XhelloXJohn"
print(greeting.split("X"))

#字串合併可迭代物件(list, tuple, set)
greeting = "++".join(data)
print(greeting)

greeting = "__".join(("WR", "#R")) #tuple
print(greeting)
```
