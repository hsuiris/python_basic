```python
# 初始化一個 list
nums = [1, 2, 3]
```

```python
# 新增元素在 list 尾端
nums = [1, 2, 3]
nums.append(4)
print(nums)
```

```python
# 修改索引位置的元素
nums = [10 ,20, 30]
nums[1] = 25
print(nums)
```

```python
# 刪除索引位置的元素
# list.pop(<index>)
nums = [10 ,20, 30]
nums.pop(1)
print(nums)
```

```python
# 新增元素 9 在指定索引 1，其餘元素往後移
# list.insert(<index>, <value>)
nums = [10 ,20, 30]
nums.insert(1,15)
print(nums)
```

```python
# 移除指定的值
# list.remove(<value>)
nums = [10 ,20, 30]
nums.remove(30)
print(nums)

#在字串後面加上5個9，由於.remove一次只會移除一個數字，所以要用搭配while
nums = [10,20,30]
for i in range(5):
    nums.append(9)
print(nums)

while 9 in nums:
    nums.remove(9)
print(nums)
```

```python
# 計算 list 內的某個值 共有幾個
nums = [10,20,30]
for i in range(5):
    nums.append(9)
print(nums)

print(f"資料有{nums.count(9)}個9")

#len() 計算list內有幾個元素
print(f"nums有{len(nums)}個元素")
```

### list 搭配 for 迴圈

```python
# 建立一個 1~10 的 list
nums = []
for i in range(1,11):
    nums.append(i)
print(nums)

# enumerate() 同時取得 index 與 value
for index, value in enumerate(nums):
   #print(index, value)

    if value == 5:
        print(index)
```
