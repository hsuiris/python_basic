```python
# 初始化 dict
fruit_price = {"apple": 50, "orange": 30, "tomato": 100}
print(fruit_price)
```

```python
# 印出 蘋果的價格
# 用key來去取value
print(fruit_price["apple"])
```

```python
# 修改 橘子的價格
fruit_price["orange"] = 200
print(fruit_price["orange"])
```

```python
# 刪除 番茄
fruit_price.pop("tomato")
print(fruit_price)
```

```python
# 建立一個 value 含有 dict 的 dict
# info為dict, phone_nums是list
user_info = {
    "name": "Iris",
    "age": 30,
    "info": {
        "nickname": "boatman",
        "role": "Pochita",
        "phone_nums": ["0915738457", "0927597329"]
    }
}
print(user_info)
```

```python
# 取得 nickname
print(user_info["info"]["nickname"])
```

```python
# 取得所有 phone_number
print(user_info["info"]["phone_nums"])

#取得phone number第一筆
print(user_info["info"]["phone_nums"][0])

```

```python
'''
改寫 fruit_dict 每個水果需要 price, quantity/amount, color
apple 有 綠色 和 紅色
orange 有 橘色
tomato 有 紅色
'''
fruit_dict = {
    "apple":{
        "price": 20,
        "quantity": 5,
        "color": ["green","red"]
    },
    "orange":{
        "price": 32,
        "quantity": 5,
        "color": ["orange"]
    },
    "tomato":{
        "price": 16,
        "quantity": 4,
        "color": ["red"]
    }
}
print(fruit_dict["apple"]["quantity"])
print(fruit_dict["apple"]["color"])
```

##### dict 的合併

```python
# dict 的合併
# 使用dict.update()

fruit_dict2 = {
    "kiwi":{
        "price": 100,
        "quantity": 10,
        "color": ["yellow","green"]
    }
}
fruit_dict.update(fruit_dict2)
print(fruit_dict)
```

##### 另外的 print 方式 `pprint` (pretty-print) 

```python
# 從 Module 引入 function
from pprint import pprint
pprint(fruit_dict)
#pprint打印出來沒有照順序，要加上sort_dict = False
pprint(fruit_dict, sort_dicts = False)
```

```python
'''
新增顏色
apple 有新的品種 顏色要有黃色
tomato 有綠色
'''

fruit_dict["apple"]["color"].append("yellow")
fruit_dict["tomato"]["color"].append("green")
print(fruit_dict)

# 如果新的顏色不在list裡面，才要新增該顏色
new_color = "green"
if new_color not in fruit_dict["kiwi"]["color"]:
    fruit_dict["kiwi"]["color"].append(new_color)

print(fruit_dict)
```
