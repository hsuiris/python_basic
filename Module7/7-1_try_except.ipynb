##### try...except...

```python
# 可以連續定義數個例外處理機制
# 最符合的例外事件類別會先拋出
try :
    x = 100
    print(y) #發生錯誤，直接跳到except
    print(x) #不會執行到

except: 
    print("發生錯誤")
```

```python
# 多層例外處理:
try :
    x = 100
    print(y) #發生錯誤
    print(x) 

#as: 賦值
except Exception as e: # except接錯誤類別 
    print(f"發生錯誤: {e}")
```

```python
try :
    x = 100
    print(x[0]) #發生錯誤
    print(y) 
    print(x) 

except NameError as e: # except 指定錯誤類別 
    print(f"發生 NameError: {e}")

except Exception as e: # except接錯誤類別 
    print(f"發生錯誤: {e}")

```
