##### 將 data.txt 的資料分類成另外三個檔案，分別為 cc.txt 、 css.txt 、 c.txt  

```python
with open(file="data.txt", mode="r", encoding="utf-8") as f:
    data = f.readlines()
print(data)

for text in data:
    text = text.strip("\n")
    if text[-2:] == "cc":
        with open(file="cc.txt", mode="a", encoding="utf-8") as f:
            f.write(text+"\n")
    elif text[-3:] == "css":
        with open(file="css.txt", mode="a", encoding="utf-8") as f:
            f.write(text+"\n")
    elif text[-1:] == "c":
        with open(file="c.txt", mode="a", encoding="utf-8") as f:
            f.write(text+"\n")
```
