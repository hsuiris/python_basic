### `json` 套件

不同資料之間做資料(只能是list, dict)的傳遞會使用json 

json.dumps()

json.loads()

```python
# 用單純的with write(), read()只能寫入string, 不能很好的操作list, dict

import json  
fruit_dict = {    
    'apple': {
        'price': 100,
        'quantity': 50,
        'color': ['green', 'red']
    },
    'orange': {
        'price': 50,
        'quantity': 30,
        'color': ['orange'] 
    }
}

# 第二種寫法
with open("fruit.json","w") as f:
    f.write(json.dumps(fruit_dict))
           # json.dumps將之轉成json字串
    
# 第二種寫法(推薦)
with open("fruit2.json","w") as f:
    json.dump(fruit_dict, f, indent=4)
    # json.dump(<list>,<要寫入的file>, 空格數量)
```

```python
# 舊有的
with open("fruit2.json","r") as f:
    data = f.read()
print(data) # 字串型態

#  load()會自動轉成該有的型別
fruit_dict = json.loads(data)
print(fruit_dict) # dict型態

# 第二種寫法
with open("fruit2.json","r") as f:
    data = json.load(f)
print(data) #dict型態
```
