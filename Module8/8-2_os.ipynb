```python
# 匯入套件
import os

# 看當前的資料夾的檔案
files = os.listdir()
print(files)
# 看指定的資料夾的檔案
folder_path = "../Module7"
files = os.listdir(folder_path)
print(files)

# 如果指定路徑的檔案存在，則執行刪除
print(os.path.exists("data.txt"))
if os.path.exists("data.txt"):
    os.remove("data.txt")
```

```python
# os.mkdir()建立資料夾
folder = "test"
if not os.path.exists(folder):
# 若沒有這個資料夾則建立
    os.mkdir(folder)
```
